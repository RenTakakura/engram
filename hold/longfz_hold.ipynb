```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
matplotlib.use('Agg')
```

```python
Index=["A31","A32","A41","A42"]

name = 'ID138464-6_%s_conv.csv'%(Index[0])

df=pd.read_csv(name, header=0)

df['phase']=Index[0]

for i in range(1,len(Index)):
    
    name = 'ID138464-6_%s_conv.csv'%(Index[i])

    df1=pd.read_csv(name, header=0)
    
    df1['phase']=Index[i]

    df=pd.concat([df,df1])
    
```

```python
df_FZ=df[df['FZ']==1]
df_FZ["session"]=0
```

```python
plt.scatter(df_FZ["X"],df_FZ["Y"],marker="o",color="blue")
```

```python
Reg=[[0,100],[120,200]]
df_FZ=df_FZ[df_FZ["X"]>Reg[0][0]]
df_FZ=df_FZ[df_FZ["X"]<Reg[0][1]]
df_FZ=df_FZ[df_FZ["Y"]>Reg[1][0]]
df_FZ=df_FZ[df_FZ["Y"]<Reg[1][1]]
```

```python
ses=df_FZ.columns.get_loc('session')

for i in range(1,df_FZ.shape[0]):
    
    if (df_FZ.iloc[i].loc["conseqFZ"]<df_FZ.iloc[i-1].loc["conseqFZ"]):
        df_FZ.iloc[i,ses]=df_FZ.iloc[i-1,ses]+1
    
    else:
    
        df_FZ.iloc[i,ses]=df_FZ.iloc[i-1,ses] 
```

```python
df_FZ
```

```python
for i in range(0,df_FZ.iloc[df_FZ.shape[0]-1,ses]):
    
    dfs =df_FZ[df_FZ['session']==i]
    
    title_name="phase %s : session %d : (X,Y)=(%d,%d)" % (dfs.iloc[0].loc["phase"],dfs.iloc[0].loc["session"],dfs.iloc[0].loc["X"],dfs.iloc[0].loc["Y"])
    
    for j in range(1,dfs.shape[0]):
    
        plt.scatter(range(1,len(dfs.iloc[j].loc['C000':'C261'])+1),dfs.iloc[j].loc['C000':'C261'],marker="o",color="red")
    
    plt.title(title_name)
    
    plt.xlim
    
    plt.show()
```

```python
df_FZc=pd.DataFrame()

df_FZc["0.005"]=(df_FZ.loc[:,"C000":"C261"] >= 0.005).sum()
df_FZc["0.05"]=(df_FZ.loc[:,"C000":"C261"] >= 0.05).sum()
df_FZc["0.02"]=(df_FZ.loc[:,"C000":"C261"] >= 0.02).sum()
df_FZc["0.10"]=(df_FZ.loc[:,"C000":"C261"] >= 0.10).sum()
df_FZc["0.15"]=(df_FZ.loc[:,"C000":"C261"] >= 0.15).sum()
df_FZc["0.20"]=(df_FZ.loc[:,"C000":"C261"] >= 0.20).sum()
```

```python
df_FZc_th=df_FZc[df_FZc["0.05"]>200]
df_FZc_th.shape[0]
```

```python
df_FZ_th0=df_FZ.loc[:,df_FZc_th.index]
df_FZc_th.index
```

```python
df_FZ_th1=df_FZ.loc[:,[df_FZc_th.index[0],"X","Y","session","phase","conseqFZ"]]
df_FZ_th=pd.merge(df_FZ_th0,df_FZ_th1, left_on=df_FZc_th.index[0], right_on=df_FZc_th.index[0])
```

```python
df_FZ_th_=df_FZ_th

for i in range(0,df_FZc_th.shape[0]):

    col_b=df_FZ_th_.columns[i]
    
    col_a=col_b[1:]
    
    df_FZ_th_ = df_FZ_th_.rename(columns={col_b:col_a}) 
    
df_FZ_th_
```

```python
for i in range(0,df_FZ_th_.iloc[df_FZ_th_.shape[0]-1].loc["session"]):
    
    dfs =df_FZ_th_[df_FZ_th_['session']==i]
    
    title_name="phase %s : session %d : (X,Y)=(%d,%d)" % (dfs.iloc[0].loc["phase"],dfs.iloc[0].loc["session"],dfs.iloc[0].loc["X"],dfs.iloc[0].loc["Y"])
    
    for j in range(1,dfs.shape[0]):
    
        plt.scatter(dfs.columns[0:df_FZc_th.shape[0]],dfs.iloc[j].loc[df_FZ_th_.columns[0]:df_FZ_th_.columns[df_FZc_th.shape[0]-1]],marker="o",color="red")
    
    plt.title(title_name)
    
    plt.ylim(0.05,0.15)
    
    plt.show()
```
