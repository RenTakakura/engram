```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Index=["A1","A2","A31","A32","A41","A42","A5","A6","A7","B1","B2","B3","B4"]

name = 'ID138464-6_%s_conv.csv'%(Index[0])

df=pd.read_csv(name, header=0)

for i in range(1,len(Index)):
    
    name = 'ID138464-6_%s_conv.csv'%(Index[i])

    df1=pd.read_csv(name, header=0)

    df=pd.concat([df,df1])
```

```python
dfev=pd.read_csv("ID138464-6_Longitudinal_Events.csv",header=0)
dfarea=df.loc[:,["X","Y","FZ","conseqFZ","Time2"] ]
```

```python
dfevents=pd.merge(dfarea,dfev, left_on='Time2', right_on='Time (s)')
```
