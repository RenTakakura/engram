```python
#import modules

#dataframe
import pandas as pd
#number processing
import numpy as np
#standard scaling
from sklearn.preprocessing import StandardScaler
#plotting
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

#heatmap
import seaborn as sns

#math
import math
```

```python
#マウスリストの作成、神経データが記録されている全フェーズリストの作成
mouse_17list=["ID170704C","ID170704E","ID170704F","ID170704G"]
mouse_13list = ["ID138464-1","ID138464-2","ID138464-3","ID138464-4","ID138464-5","ID138464-6"]
mouse_list=mouse_13list+mouse_17list
phaseall_list=["HC1","B1","B2","HC2","A1","A2","A31","A32","A41","A42","HC3","A5","B3","A6","HC4","A7","B4"]
phase_Alist=["A1","A20","A21","A22","A23","A31","A32","A41","A42","A5","A60","A61","A62","A63","A7"]
phase_Blist=["B1","B2","B3","B4"]
phase_Hlist=["HC1","HC2","HC3","HC4"]
phase_ABlist=phase_Alist+phase_Blist
phase_shocklist=["A2","A6"]
phase_shockdivlist=[0,1,2,3]
phase_daydivlist=[0,3,10,14,17]
```

```python
#　分析実行
for m in range(0,len(mouse_list)):

    mouse = mouse_list[m]
    
    #神経データをデータフレームにインポートする・Accepted cellのみ抽出する
    inputf='%s/python/%s_Longitudinal_Traces.csv' %(mouse,mouse)
    dff=pd.read_csv(inputf,header=0)
    dff0=pd.DataFrame()
    dff0=dff.iloc[0,:]
    Accept=dff0.index[dff0==' accepted']
    dfa=pd.DataFrame()
    dfa['Time2']=dff.iloc[1:,0]
    for accept_c in range (0,len(Accept)):
        dfa[Accept[accept_c]]=dff[Accept[accept_c]].iloc[1:]
        dfa=dfa.astype('float')

    #Acceped tcellのfirst cellとlast cellを抽出
    cell_f=dfa.columns[1]
    cell_l=dfa.columns[dfa.shape[1]-1]
    
    for ph in range(0,len(phase_Blist)):
        
        phase = phase_Blist[ph]
        
        inputf='%s/python/%s_%s_mean_std.csv' %(mouse,mouse,phase)
        dfmean=pd.read_csv(inputf,header=0)
        
        #mouseの運動軌道 :trajectory(X:x,Y:y)を作成
        plt.figure()
        plt.figure(figsize=(5.4, 3.6), dpi=80)
        plt.subplots_adjust(right=0.75)
        plt.ylim(0,200)
        plt.xlim(0,250)
        plt.plot(dfmean["X"],dfmean["Y"],marker="o",color="black")
        df_FZ=dfmean[dfmean["FZ"]!=0]
        df_nFZ=dfmean[dfmean["FZ"]==0]
        title_name='trajectory_%s' % (phase)
        plt.title(title_name)
        file_name='%s/python/%s_figure/'%(mouse,mouse)+title_name+'.png'
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.savefig(file_name)
        plt.figure()

        #mouseのfreezing挙動を示した記録:FZpoint(X:x,Y:y)を作成
        plt.figure()
        plt.figure(figsize=(5.4, 3.6), dpi=80)
        plt.subplots_adjust(right=0.75)
        plt.scatter(df_FZ["X"],df_FZ["Y"],marker="o",color="black")
        plt.ylim(0,200)
        plt.xlim(0,250)
        title_name='FZpoint_%s' % (phase)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title(title_name)
        file_name='%s/python/%s_figure/'%(mouse,mouse)+title_name+'.png'
        plt.savefig(file_name)
        plt.figure()
        
        #shockを与えるフェーズにおいて、mouseがshockを与えられた時に位置していた場所を
        #滞在時刻とともに記録:shock_point(X:x,Y:y,annotate:slice)を作成
        if phase=='A2' or phase=='A6':
            plt.figure()
            plt.figure(figsize=(5.4, 3.6), dpi=80)
            plt.subplots_adjust(right=0.75)
            df_shock=dfmean[dfmean['Shock']==1]
            plt.scatter(df_shock["X"],df_shock["Y"],marker="o",color="red")
            for i in range(0,df_shock.shape[0]):
                plt.annotate(int(df_shock.iloc[i].loc["Slice"]), (df_shock.iloc[i].loc["X"], df_shock.iloc[i].loc["Y"]),size=10)
            plt.xlim(0,250)
            plt.ylim(0,200)
            title_name='shock_point_%s' % (phase)
            plt.title(title_name)
            file_name='%s/python/%s_figure/'%(mouse,mouse)+title_name+'.png'
            plt.savefig(file_name)
            plt.figure()
```
