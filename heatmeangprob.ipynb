ヒートマップの作成(heatmeangprob)<br>
<方針><br>
・フェーズごとと全フェーズ合算の二種類で行う<br>
・各cell各フェーズにおいて、standard(標準化)を使う<br>
・標準化後の値が2.0以上のものを「活動状態」の判定基準とする<br>
・各cellのcolorが示す値は、そのエリアに存在した時に、「活動状態」だった確率

```python
#import modules

#dataframe
import pandas as pd

#number processing
import numpy as np

#standard scaling
from sklearn.preprocessing import StandardScaler

#plotting
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

#heatmap
import seaborn as sns

#math
import math
```

```python
#make the list

#mouse
mouse_list = ["ID138464-1","ID138464-2","ID138464-3","ID138464-4","ID138464-5","ID138464-6"]
#all phase 
phaseall_list=["HC1","B1","B2","HC2","A1","A2","A31","A32","A41","A42","HC3","A5","B3","A6","A7","HC4","B4"]
#phase targeted for anlysis
phase_list=["A1","A2","A31","A32","A41","A42","A5","A6","A7"]
```

現在のバージョン(phaseごと)

```python
#分析実行


for m in range(0,6):

    mouse = mouse_list[m]
    
    #神経データをデータフレームにインポートする
    inputf='%s/python/%s_Longitudinal_Traces.csv' %(mouse,mouse)
    dff=pd.read_csv(inputf,header=0)
    dff0=pd.DataFrame()

    #Acceptのみを抽出したデータフレームの作成
    dff0=dff.iloc[0,:]
    Accept=dff0.index[dff0==' accepted']
    dfa=pd.DataFrame()
    dfa['Time2']=dff.iloc[1:,0]
    for accept_c in range (0,len(Accept)):
        dfa[Accept[accept_c]]=dff[Accept[accept_c]].iloc[1:]
        dfa=dfa.astype('float')

    #Acceptcellのfirst cellとlast cellを抽出
    cell_f=dfa.columns[1]
    cell_l=dfa.columns[dfa.shape[1]-1]
    
    #フェーズごと
    #ケージAについて標準化処理済みのデータについてフェーズを全て結合して、ヒートマップ:heatmeanAbrgprob(X:x,Y:y,heat:p(訪れた時にdiff?>ts))の作成を作成
    #スライスごとの平均値でダウンサンプリング処理済み
    #一度もマウスがvisitしなかった場所はblueで示し、visitを一回以上したマスに対して赤色系統でheatmap作成
    #gaussian kernelを使用:exp(-(1マスを単位とした距離の二乗))を前後左右2マス未満に適用(周辺8個)
    #tsは、最適なものは2
    inputf='%s/python/%s_meanstdall.csv' %(mouse,mouse)          
    dfmeanstd_all=pd.read_csv(inputf,header=0)
    dfmap=pd.DataFrame(index=np.arange(195,-5,-200/10),columns=np.arange(6.25,256.25,250/10))
    dfmap_d=pd.DataFrame(index=np.arange(195,-5,-200/10),columns=np.arange(6.25,256.25,250/10))
    dfmap_n=pd.DataFrame(index=np.arange(195,-5,-200/10),columns=np.arange(6.25,256.25,250/10))
    

    cnt=0

    for i in range(dfmeanstd_all.shape[0]):

        if (dfmeanstd_all['Time2'].iloc[i]-dfmeanstd_all['Time2'].iloc[i-1]>100):

            cnt = cnt + 1

        dfmeanstd_all['phase'].iloc[i]=phase_list[cnt]
        
    for ph in range(len(phase_list)):

        phase=phase_list[ph]

        dfmeanstd_phase=dfmeanstd_all[dfmeanstd_all['phase']==phase]


        for i in range(0,10):

            for j in range(0,10):

                n=dfmeanstd_phase[(dfmeanstd_phase['X']>float(dfmap.columns[i]-12.5) )& (dfmeanstd_phase['X']<float(dfmap.columns[i]+12.5) )&(dfmeanstd_phase['Y']>float(dfmap.index[j]-10) )& (dfmeanstd_phase['Y']<float(dfmap.index[j]+10))].shape[0]

                dfmap_n.iloc[i,j]=n
                
                
        dfmap_norm=pd.DataFrame(index=np.arange(195,-5,-200/10),columns=np.arange(6.25,256.25,250/10))

        for i in range(0,dfmap_norm.shape[0]):
        
            for j in range(0,dfmap_norm.shape[1]):
        
                dfmap_norm.iloc[i,j]=0
            
                if (dfmap_n.iloc[i,j]<0.5):
                
                    dfmap_norm.iloc[i,j]=1
                
                else:
        
                    #for ii in range(0,dfmap_norm.shape[0]):
                    for ii in range(max(0,i-1),min(dfmap_norm.shape[0],i+2)):
                
                        #for jj in range(0,dfmap_norm.shape[1]):
                        for jj in range(max(0,j-1),min(dfmap_norm.shape[1],j+2)):
                    
                            if (dfmap_n.iloc[ii,jj]>0.5):
                
                                dfmap_norm.iloc[i,j]+=math.exp(-(math.pow(i-ii,2.0)+math.pow(j-jj,2.0)))
    

        for ts in range(2,3):

            for cell in range(dfmeanstd_phase.columns.get_loc(cell_f),dfmeanstd_phase.columns.get_loc(cell_l)+1):

                dfmap=pd.DataFrame(index=np.arange(195,-5,-200/10),columns=np.arange(6.25,256.25,250/10))
                dfmap_d=pd.DataFrame(index=np.arange(195,-5,-200/10),columns=np.arange(6.25,256.25,250/10))

                for i in range(0,10):

                    for j in range(0,10):

                        if dfmap_n.iloc[i,j]<0.5:


                            dfmap.iloc[i,j]=-1

                            dfmap_d.iloc[i,j]=0

                        else:

                            d=dfmeanstd_phase[(dfmeanstd_phase['X']>float(dfmap.columns[i]-12.5)) & (dfmeanstd_phase['X']<float(dfmap.columns[i]+12.5)) & (dfmeanstd_phase['Y']>float(dfmap.index[j]-10)) & (dfmeanstd_phase['Y']<float(dfmap.index[j]+10) )& (dfmeanstd_phase[dfmeanstd_phase.columns[cell]]>ts)].shape[0]


                            dfmap_d.iloc[i,j]=d

                            dfmap.iloc[i,j]=(d/dfmap_n.iloc[i,j])

                dfmap_g=pd.DataFrame(index=np.arange(195,-5,-200/10),columns=np.arange(6.25,256.25,250/10))

                for i in range(0,dfmap_g.shape[0]):

                    for j in range(0,dfmap_g.shape[1]):

                        if (dfmap.iloc[i,j]==-1):
                        
                            dfmap_g.iloc[i,j]=-0.5
                        
                        else:
                    
        
                            dfmap_g.iloc[i,j]=0
        
                            for ii in range(max(0,i-1),min(dfmap_norm.shape[0],i+2)):
            
                                for jj in range(max(0,j-1),min(dfmap_norm.shape[1],j+2)):
                    
                                    if(dfmap.iloc[ii,jj]>-1):
                
                                        dfmap_g.iloc[i,j]+=math.exp(-(math.pow(i-ii,2.0)+math.pow(j-jj,2.0)))*dfmap.iloc[ii,jj]

                dfmap_g_norm=dfmap_g/dfmap_norm
                
                dfmap_g_normT=dfmap_g_norm.T

                dfmap_g_normT=dfmap_g_normT.astype(float)
                #dfmap.columns.astype(float)
                #dfmap.index.astype(float)

                plt.figure()
                title_name='heatmeanAbrgphaseprob_ts%s_%s_%s' % (ts,dfmeanstd_phase.columns[cell],phase)
                file_name='%s/python/%s_fig/'%(mouse,mouse)+title_name+'.png'
                plt.title(title_name)
                plt.xlabel('x')
                plt.ylabel('y')
                sns.heatmap(dfmap_g_normT,vmin=-0.1,vmax=0.3,center=0.0,cmap='seismic',xticklabels=np.arange(25,250+1,25),yticklabels=np.arange(200,19,-20))
                plt.savefig(file_name)
```
