```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

mouse_list = ["ID138464-1","ID138464-2","ID138464-3","ID138464-4","ID138464-5","ID138464-6"]

phase_list=["A1","A2","A31","A32","A41","A42","A5","A6","A7"]

mouse = mouse_list[5]

phase=phase_list[8]

file = 'ID138464-%s_%s_conv.csv' % (mouse,phase)

df = pd.read_csv(file, header=0)
```

```python
df['FZ1']=df['FZ'].shift(-10)
#df['FZ2']=df['FZ'].shift(-20)
df['FZr']=df['FZ']+df['FZ1']
df["session"]=0
```

```python
ses=df.columns.get_loc('session')

for i in range(1,df.shape[0]):
    
    if ((df.iloc[i].loc["FZr"]*df.iloc[i-1].loc["FZr"]==0) and (df.iloc[i].loc["FZr"]+df.iloc[i-1].loc["FZr"]!=0)):
        
        df.iloc[i,ses]= df.iloc[i-1,ses]+1
    
    else:
        
        df.iloc[i,ses]= df.iloc[i-1,ses]
```

```python
df_=df

for i in range(df_.columns.get_loc('C000'),df_.columns.get_loc('C261')+1):

    col_b=df_.columns[i]
    
    col_a=col_b[1:]
    
    df_ = df_.rename(columns={col_b:col_a})
```

```python
plt.plot(df_["X"],df_["Y"],marker="o",color="blue")
plt.xlim(0,250)
plt.ylim(0,200)
df_FZ=df_[df_["FZr"]!=0]
df_nFZ=df_[df_["FZr"]==0]
title_name='fig1/phase%s_trajectory_mouse%s_data%d' % (phase,mouse,df_.shape[0])
plt.title(title_name)
file_name=title_name+'.png'
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig(file_name)
plt.figure()
#plt.show()
```
